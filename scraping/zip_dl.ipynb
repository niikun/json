```python
import json, time, os, requests
from bs4 import BeautifulSoup
```

```python
baseurl = "https://api.aoikujira.com/zip/list.php"
res = requests.get(baseurl)
soup = BeautifulSoup(res.text,"html.parser")
```

```python
pref_datas = []
city_datas = []
```

```python
prefs = soup.find_all("li")
for pref in prefs:

    pref_data = pref.text.split(" ")[0]
    pref_datas.append(pref_data)

print(pref_datas)
```

```python
for pref_data in pref_datas:
    pref_url = f"https://api.aoikujira.com/zip/list.php?m=ken&ken={pref_data}"
    res_pref = requests.get(pref_url).text
    time.sleep(1)
    soup_pref = BeautifulSoup(res_pref,"html.parser")
    cities = soup_pref.find_all("li")
    for city in cities:
        city_name = city.text
        city_data = {
            "pref_name":pref_data,
            "city_name":city_name
        }
        city_datas.append(city_data)

print(city_datas)
```

```python
with open("datas.json","w", encoding="utf-8") as f:
    json.dump(city_datas,f, ensure_ascii=False, indent=2)
```
